```python
import os

# Define the directory where files will be saved
output_dir = "moved_files/documents/"
os.makedirs(output_dir, exist_ok=True)

# Define sample file content
samples = {
    "file1.txt": "ID|Name|Age\n1|Alex|30\n2|Joe|25\n3|Charlie|35",
    "file2.txt": "ID|Name|Age|Email\n1|Alex|30|alex@example.com\n2|Joe|25|Joe@example.com\n3|Charlie|35|charlie@example.com",
    "file3.txt": "Name|ID|Salary\nAlex|1|50000\Joe|2|55000\nCharlie|3|60000"
}

# Write files
for filename, content in samples.items():
    with open(os.path.join(output_dir, filename), "w") as f:
        f.write(content)

print("Sample files generated in:", output_dir)
```
